```python
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
import numpy as np
import librosa
import os
import IPython.display as ipd
```

```python
PATH = 'external_media/'
FILE = 'piano.wav'
audio_path = PATH+FILE
```

```python
wave, sr = librosa.load(audio_path, sr=22050)
print(type(wave),type(sr))
print(wave.shape, sr)

#This returns an audio time series as a numpy array with a default sampling rate(sr) of 22KHZ mono. We can change this behaviour by saying:
#* librosa.load(audio_path, sr=44100)
```

```python
ipd.Audio(audio_path)
```

## WAVEFORM PLOT

```python
plt.figure(figsize=(10, 3))
librosa.display.waveplot(wave, sr)
```

## SPECTOGRAM PLOT

```python
X = librosa.stft(wave)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
# librosa.output.write_wav('example.wav', x, sr) # WRITING OUTPUT AUDIO
```

```python
mfccs = librosa.feature.mfcc(wave, sr)
print(mfccs.shape)
```

```python
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
```

```python
import sklearn
mfccs = sklearn.preprocessing.scale(mfccs, axis=1)
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
```
